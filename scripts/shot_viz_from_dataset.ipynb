# Shot Visualization from `shot_contest_dataset.csv`

This notebook mirrors the style of `vizualisation.ipynb`, but drives visuals from shot-level rows.

- Uses `game_file`, `game_id`, `release_frame`, `shooter_id` from `shot_contest_dataset.csv`
- Loads full Hawk-Eye parquet from OneDrive
- Produces:
  1. Release-frame 3D image (`release_snapshot.png`)
  2. Animation window (seconds before release to a few frames after)


```python
import os
import pandas as pd
from shot_viz_from_dataset import process_one_shot
```

```python
# Configure your paths
SHOTS_CSV = "../data/outputs/shot_contest_dataset.csv"
PARQUET_DIR = "/Users/mariaangellobon/Library/CloudStorage/OneDrive-SharedLibraries-MassachusettsInstituteofTechnology/[MIT] Basketball Officiating - miami_heat_2025"
OUTPUT_DIR = "../data/outputs/shot_visualizations"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Animation window around release
PRE_FRAMES = 120  # ~2s at 60Hz
POST_FRAMES = 6
```

```python
# Preview shots table
shots = pd.read_csv(SHOTS_CSV)
print(f"rows: {len(shots)}")
shots.head(3)
```

```python
# Single-shot example (1-based row index from CSV display)
ROW_INDEX = 150
shot_row = shots.iloc[ROW_INDEX - 1]

record = process_one_shot(
    shot_row,
    parquet_dir=PARQUET_DIR,
    output_dir=OUTPUT_DIR,
    pre_frames=PRE_FRAMES,
    post_frames=POST_FRAMES,
)
record
```

```python
# Batch example: run first N shots (adjust as needed)
N = 5
records = []
for _, row in shots.head(N).iterrows():
    try:
        rec = process_one_shot(
            row,
            parquet_dir=PARQUET_DIR,
            output_dir=OUTPUT_DIR,
            pre_frames=PRE_FRAMES,
            post_frames=POST_FRAMES,
        )
        records.append(rec)
    except Exception as exc:
        print("FAIL", row.get("game_id"), row.get("release_frame"), row.get("shooter_id"), exc)

pd.DataFrame(records)
```
